```python

path_dir = 'template'
templates = os.listdir(path_dir)

template_dictionary = dict()
    
for temp, point in zip(templates[:len(points)], points):
    print(temp, point.shape)
    img = cv2.imread(os.path.join(path_dir, temp))
    img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    orb = cv2.ORB_create()
    kp, des = orb.detectAndCompute(img_gray, None)

    

    #      
    
#     template_dictionary.append([kp, des])

    pt = [elem.pt for elem in kp]
    size = [elem.size for elem in kp]
    angle = [elem.angle for elem in kp]
    response = [elem.response for elem in kp]
    octave = [elem.octave for elem in kp]
    class_id = [elem.class_id for elem in kp]
    
    template_dictionary[temp] = dict({'points': point, 'kp': (pt, size, angle, response, octave, class_id)  , 'des': des})


    points1 = np.float32(point[0]).reshape(-1, 1, 2)
    points2 = np.float32(point[1]).reshape(-1, 1, 2)
    points3 = np.float32(point[2]).reshape(-1, 1, 2)
    cv2.polylines(img_gray, [np.int32(points1)], True, 0,5, cv2.LINE_AA)
    cv2.polylines(img_gray, [np.int32(points2)], True, 0,5, cv2.LINE_AA)
    cv2.polylines(img_gray, [np.int32(points3)], True, 0,5, cv2.LINE_AA)

#     cur_points = cv2.drawKeypoints(img, kp, img, color=255)

    plt.imshow(img_gray, cmap='gray')
    plt.show()
```
